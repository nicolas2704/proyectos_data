# `**dataset Iris **`
El conjunto de datos de Iris contiene cuatro caracter√≠sticas (longitud y anchura de s√©palos y p√©talos) de 50 muestras de tres especies de flores Iris (Iris setosa, Iris virginica e Iris versicolor).

# `**dataset Penguins **`
El dataset palmer penguins es un conjunto de datos sobre ping√ºinos de tres especies diferentes que anidan en una colonia de Archipi√©lago Palmer, en la Ant√°rtida.
Contiene informaci√≥n sobre las medidas corporales de los ping√ºinos, incluyendo la longitud y profundidad del pico, la longitud de la aleta y la masa corporal, as√≠ como informaci√≥n sobre el sexo y la edad de los ping√ºinos.

Hay tres especies de ping√ºinos en el conjunto de datos: Adelie, Gentoo, y Chinstrap.

```python
import plotly.express as px  # Importa la librer√≠a Plotly Express para crear gr√°ficos interactivos.
import seaborn as sns        # Importa la librer√≠a Seaborn para cargar conjuntos de datos utilizados.
import pandas as pd          # Importa Pandas para la manipulaci√≥n y an√°lisis de datos.

# Se define una funci√≥n plot_interactive_graphs que permite generar graficos interactivos
def plot_interactive_graphs():
    # Solicitamos al usuario que elija uno de los conjuntos de datos disponibles, "iris" o "palmer_penguins" >> "palmer" para su an√°lisis
    dataset_name = input("¬øQu√© conjunto de datos desea utilizar? ( iris / penguins ): ").strip().lower()

    # Carga el conjunto de datos en funci√≥n de la selecci√≥n del usuario
    if dataset_name == 'iris':
        data = sns.load_dataset('iris')  # Carga el conjunto de datos 'iris' utilizando Seaborn.
        print(" ** Estructura del conjunto de datos seleccionado ** ")
        print(data.columns)
        data=data.rename(columns={"sepal_length":"largo_sepalo","sepal_width":"ancho_sepalo","petal_length":"largo_petalo",
                                  "petal_width":"ancho_petalo","species":"especies"})
        print(data.head(5))  # Muestra las primeras filas del conjunto de datos para una inspecci√≥n inicial.

        print(" ** Distribuci√≥n x tipo de flores iris ** ")
        print(data['especies'].value_counts()) # Cantidad de datos x  especies

       # Agrupa los datos por la columna 'species' y calcula el promedio de 'sepal_length' para cada tipo de iris.
        data_bar = data.groupby('especies', as_index=False)['largo_sepalo'].mean()


        bar_title = 'Gr√°fico de Barras: Longitud promedio del s√©palo (cm)'  # Define el t√≠tulo para el gr√°fico de barras.
        box_title = 'Gr√°fico de Bigotes: Anchura del s√©palo (cm)'  # Define el t√≠tulo para el gr√°fico de bigotes.
        pie_title = 'Gr√°fico Circular: Distribuci√≥n de Tipo de Flores'  # Define el t√≠tulo para el gr√°fico Circular.

        x_column = 'especies'  # Define la columna que se usar√° en el eje x, en este caso, 'species' (especies).
        y_column_bar = 'largo_sepalo'  # Define la columna que se usar√° en el eje y del gr√°fico de barras, 'sepal_length' (largo del s√©palo).
        y_column_box = 'ancho_sepalo'  # Define la columna que se usar√° en el eje y del gr√°fico de barras, 'sepal_width' (anchura del s√©palo).

        color_column = 'especies'  # Define la columna que se usar√° para colorear los datos en los gr√°ficos, en este caso, 'species'.

    elif dataset_name == 'penguins':
        data = sns.load_dataset('penguins')  # Carga el conjunto de datos 'penguins' utilizando Seaborn.

        print(" ** Estructura del conjunto de datos seleccionado ** ")
        data=data.rename(columns={"species":"especies","island":"islas","bill_length_mm":"largo_pico_mm","bill_depth_mm":"profundidad_pico_mm",
                                  "flipper_length_mm":"largo_aletas_mm","body_mass_g":"masa_corporal_gr","sex":"sexo"})
        print(data.head(5))  # Muestra las primeras filas del conjunto de datos para una inspecci√≥n inicial.

        print(" ** Distribuci√≥n x tipo de pinguinos ** ")
        print(data['especies'].value_counts()) # Cantidad de datos x especies

        # Agrupa los datos por la columna 'species' y calcula el promedio de 'flipper_length_mm' para cada tipo de corte.
        data_bar = data.groupby('especies', as_index=False)['largo_aletas_mm'].mean()

        bar_title = 'Gr√°fico de Barras: Longitud promedio de las aletas (mm)'  # T√≠tulo para el gr√°fico de barras.
        box_title = 'Gr√°fico de Bigotes: Profundidad del pico (mm)'  # T√≠tulo para el gr√°fico de bigotes.
        pie_title = 'Gr√°fico Circular: Distribuci√≥n de Especies de Pinguinos'  # Define el t√≠tulo para el gr√°fico Circular.

        x_column = 'especies'  # Define la columna que se usar√° en el eje x, en este caso, 'species' (tipo de pinguinos).
        y_column_bar = 'largo_aletas_mm'  # Define la columna para el eje y del gr√°fico de barras, 'flipper_lenght_mm' (largo de las aletas).
        y_column_box = 'profundidad_pico_mm'  # Define la columna para el eje y del gr√°fico de barras, 'bill_depth_mm' (profundidad del pico).
        color_column = 'especies'  # Define la columna para colorear los datos en los gr√°ficos, en este caso, 'species'.

    else:
         # Muestra un mensaje de error si el usuario no selecciona un conjunto de datos v√°lido y finaliza la funci√≥n.
        print("Error: el nombre del dataset debe ser 'iris' o 'penguins'.")
        return

    # Gr√°fico de barras (bar): muestra el promedio del largo de la especie
    fig_bar = px.bar(data_bar, x=x_column, y=y_column_bar, title=bar_title, color=color_column,
                      color_discrete_map={'Especie1': '#C65AD6',
                                          'Especie2': '#50B5AE',
                                          'Especie3': '#E07924'})
    fig_bar.update_layout(template='plotly_dark')  # Aplicar el tema oscuro

    # Gr√°fico circular (pie): muestra la distribuci√≥n de la columna seleccionada.
    fig_pie = px.pie(data, names=x_column, title=pie_title,
                     color=x_column,
                     color_discrete_map={'Especie1': '#C65AD6',
                                          'Especie2': '#50B5AE',
                                          'Especie3': '#E07924'})
    fig_pie.update_layout(template='plotly_dark')  # Aplicar el tema oscuro

    # Gr√°fico de bigotes (boxplot): muestra la distribuci√≥n de la columna
    fig_box = px.box(data, x=x_column, y=y_column_box, title=box_title,
                     color=x_column,
                     color_discrete_map={'Especie1': '#C65AD6',
                                          'Especie2': '#50B5AE',
                                          'Especie3': '#E07924'})
    fig_box.update_layout(template='plotly_dark')  # Aplicar el tema oscuro

    fig_bar.show()  # Muestra el gr√°fico de barras.
    fig_pie.show()  # Muestra el gr√°fico circular.
    fig_box.show()  # Muestra el gr√°fico de bigotes.

# Ejecutar la funci√≥n para solicitar el dataset y generar los gr√°ficos
plot_interactive_graphs()
```
